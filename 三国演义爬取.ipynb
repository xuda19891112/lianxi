```python
import requests
import urllib
import re
import os
import time
from bs4 import BeautifulSoup
```

```python
mainurl = 'https://www.shicimingju.com/book/sanguoyanyi.html'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'
}

# 获取主页内容并解码
response = requests.get(url = mainurl, headers=headers)
response.encoding = ('utf-8')
res_text = response.text

# 获取章节名称及url
titleList = []
titleUrl = []
soup = BeautifulSoup(res_text,'lxml')
title_list = soup.select('.book-mulu > ul > li > a')
for i in title_list:
    title = i.text
    title_url = 'https://www.shicimingju.com' + i['href']
    titleList.append(title)
    titleUrl.append(title_url)
    print(title,':',title_url)
print(titleList)
print(titleUrl)

# 循环遍历每个章节
num = 0
with open('三国演义.txt','w+',encoding='utf-8') as fp:
    for url in titleUrl:
        response1 = requests.get(url = url, headers=headers)
        response1.encoding = "utf-8"
        res1_text = response1.text
        soup1 = BeautifulSoup(res1_text,'lxml')

        titleTemp = titleList[num]
        divtag = soup1.find('div',class_="chapter_content")
        contentTemp = divtag.text
        fp.write(titleTemp+'\n'+contentTemp+'\n')
        print(title,"保存成功！")
        num = num+1

with open('1.jgp','wb') as fp:
    fp.write(img_data)
```
