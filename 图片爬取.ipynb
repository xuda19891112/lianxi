```python
from bs4 import BeautifulSoup
import requests
import re
import os
from  lxml import etree
# import selenium
print(os.getcwd())
dirName = 'girlslib'
if not os.path.exists('girlslib'):
    os.mkdir('girlslib')
url1 = 'https://pic.netbian.com/4kmeinv/'
# url2 = 'https://pic.netbian.com/4kmeinv/index_%d.html'%a
headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36'
}

for a in range(1,6):
    if a ==1:
        url = url1
    else:
        url = 'https://pic.netbian.com/4kmeinv/index_%d.html'%a
    page_response = requests.get(url = url,headers=headers)
    page_response.encoding = 'gbk'
    page_text = page_response.text
    tree = etree.HTML(page_text) 
    li_list = tree.xpath('//div[@class="slist"]//li')
    for i in li_list:
        title = i.xpath('.//img/@alt')[0] + '.jpg'
        img_url = 'https://pic.netbian.com'+i.xpath('.//img/@src')[0]
        print(title,img_url)
        img_response = requests.get(url = img_url,headers=headers).content
        img_path = dirName + '/'+ title
        with open(img_path,'wb') as fp:
            fp.write(img_response)


```
